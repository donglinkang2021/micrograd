```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## single value

```python
# def f(x):
#     return 3*x**2 - 4*x + 5

# use lambda function to define f(x)
f = lambda x: 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
plt.figure(figsize=(5, 3))
plt.plot(xs, ys)    
plt.grid()
plt.show()
```

```python
h = 1e-8
df = lambda x: (f(x+h) - f(x))/h
df(3.0)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, f(xs))    
plt.plot(xs, df(xs))
plt.grid()
plt.show()
```

## multiple values

```python
g = lambda a,b,c : a * b + c
```

```python
a = 2.0
b = -3.0
c = 10.0
g(a,b,c)
```

```python
h = 1e-8
dg_da = lambda a, b, c: (g(a+h, b, c) - g(a, b, c))/h
dg_db = lambda a, b, c: (g(a, b+h, c) - g(a, b, c))/h
dg_dc = lambda a, b, c: (g(a, b, c+h) - g(a, b, c))/h

dg_da(a, b, c), dg_db(a, b, c), dg_dc(a, b, c)
```

```python
# %matplotlib widget
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

a = np.arange(-5, 5, 0.25)
b = np.arange(-5, 5, 0.25)
a, b = np.meshgrid(a, b)
z = g(a, b, 10.0)
da = dg_da(a, b, 10.0)
db = dg_db(a, b, 10.0)


ax.plot_surface(a, b, z, cmap='coolwarm')
ax.quiver(a, b, z, da, db, 0, length=0.5, normalize=True, color='black')
plt.show()
```
